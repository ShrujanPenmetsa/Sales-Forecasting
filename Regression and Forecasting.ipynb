```python
import pandas as pd

file_path = 'System2_Restaurant Case Study Data.xlsx'
xls = pd.ExcelFile(file_path)
reported_numbers_df = pd.read_excel(xls, sheet_name='Reported Numbers')

txrh_data = reported_numbers_df[reported_numbers_df['company'] == 'Texas Roadhouse Inc.']
eat_data = reported_numbers_df[reported_numbers_df['company'] == 'Brinker International, Inc.']
dri_data = reported_numbers_df[reported_numbers_df['company'] == 'Darden Restaurants, Inc.']

#adding earlier dates as they would be helpful in YOY calculations
# Dates to add for TXRH
txrh_new_dates = ['9/1/16', '9/26/16', '12/26/16', '3/27/17', '6/26/17']
txrh_new_rows = pd.DataFrame({
    'entity': ['TXRH'] * len(txrh_new_dates),
    'company': ['Texas Roadhouse Inc.'] * len(txrh_new_dates),
    'metric': txrh_data['metric'].iloc[0] if not txrh_data.empty else 'sales_yoy',  # Default metric if no data exists
    'period_end_dt': pd.to_datetime(txrh_new_dates),
    'reported_sales_yoy': [None] * len(txrh_new_dates)  # Placeholder for new rows
})

# Dates to add for EAT
eat_new_dates = ['9/1/16', '9/26/16', '12/27/16', '3/28/17', '6/27/17']
eat_new_rows = pd.DataFrame({
    'entity': ['EAT'] * len(eat_new_dates),
    'company': ['Brinker International, Inc.'] * len(eat_new_dates),
    'metric': eat_data['metric'].iloc[0] if not eat_data.empty else 'sales_yoy',
    'period_end_dt': pd.to_datetime(eat_new_dates),
    'reported_sales_yoy': [None] * len(eat_new_dates)
})

# Dates to add for DRI
dri_new_dates = ['9/1/16', '11/26/16', '2/25/17']
dri_new_rows = pd.DataFrame({
    'entity': ['DRI'] * len(dri_new_dates),
    'company': ['Darden Restaurants, Inc.'] * len(dri_new_dates),
    'metric': dri_data['metric'].iloc[0] if not dri_data.empty else 'sales_yoy',
    'period_end_dt': pd.to_datetime(dri_new_dates),
    'reported_sales_yoy': [None] * len(dri_new_dates)
})
updated_reported_numbers_df = pd.concat([txrh_new_rows, eat_new_rows, dri_new_rows, reported_numbers_df])
updated_reported_numbers_df = updated_reported_numbers_df.sort_values(by='period_end_dt').reset_index(drop=True)
updated_reported_numbers_df_sorted = updated_reported_numbers_df.sort_values(by=['company', 'period_end_dt']).reset_index(drop=True)
updated_reported_numbers_df_sorted
```

```python
import pandas as pd

txn_data_new = pd.read_excel("System2_Restaurant Case Study Data.xlsx", sheet_name='Txn Data')
panel_stats = pd.read_excel("System2_Restaurant Case Study Data.xlsx", sheet_name='Panel Stats')

panel_stats['optimized_transaction_date'] = pd.to_datetime(panel_stats['optimized_transaction_date'])
txn_data_new['optimized_transaction_date'] = pd.to_datetime(txn_data_new['optimized_transaction_date'])
merged_data = pd.merge(txn_data_new, panel_stats, on='optimized_transaction_date', how='left')
reported_numbers = updated_reported_numbers_df_sorted
reported_numbers['period_end_dt'] = pd.to_datetime(reported_numbers['period_end_dt'])

reported_numbers_sorted = reported_numbers.sort_values(by=['entity', 'period_end_dt'])
```

```python
#Aggregating spend, transaction and panel using sum for each period in the reported numbers sheet

reported_numbers_sorted['total_captured_spend'] = 0.0
reported_numbers_sorted['total_num_txns'] = 0
reported_numbers_sorted['total_num_panelists'] = 0

for entity in reported_numbers_sorted['entity'].unique():
    # Filter reported numbers for the current entity
    entity_reported_numbers = reported_numbers_sorted[reported_numbers_sorted['entity'] == entity]
    
    # Get the merged data specific to this restaurant
    entity_merged_data = merged_data[merged_data['entity'] == entity]
    
    for i in range(1, len(entity_reported_numbers)):
        # Get the start and end dates for the current period
        start_date = entity_reported_numbers.iloc[i - 1]['period_end_dt']
        end_date = entity_reported_numbers.iloc[i]['period_end_dt']
        
        # Filter the merged data for the date range between start_date and end_date
        date_filtered_data = entity_merged_data[
            (entity_merged_data['optimized_transaction_date'] > start_date) & 
            (entity_merged_data['optimized_transaction_date'] <= end_date)
        ]
        
        # Aggregate the spend, transactions, and panelists for this period
        total_captured_spend = date_filtered_data['captured_spend'].sum()
        total_num_txns = date_filtered_data['num_txns'].sum()
        total_num_panelists = date_filtered_data['num_panelists'].sum()
        
        # Update the reported_numbers_sorted with the calculated values
        reported_numbers_sorted.loc[(reported_numbers_sorted['entity'] == entity) & 
                                    (reported_numbers_sorted['period_end_dt'] == end_date),
                                    ['total_captured_spend', 'total_num_txns', 'total_num_panelists']] = [
            total_captured_spend, total_num_txns, total_num_panelists
        ]
```

```python
reported_numbers_sorted
```

```python
# Calculate YoY change for spend_per_panelist, txn_per_panelist, and spend_per_txn 
reported_numbers_sorted['spend_per_panelist'] = reported_numbers_sorted['total_captured_spend'] / reported_numbers_sorted['total_num_panelists']
reported_numbers_sorted['txn_per_panelist'] = reported_numbers_sorted['total_num_txns'] / reported_numbers_sorted['total_num_panelists']

# Add a new column for total_captured_spend/total_num_txns (spend per transaction)
reported_numbers_sorted['spend_per_txn'] = reported_numbers_sorted['total_captured_spend'] / reported_numbers_sorted['total_num_txns']

reported_numbers_sorted['spend_per_panelist_yoy'] = reported_numbers_sorted.groupby('entity')['spend_per_panelist'].pct_change(periods=4)*100
reported_numbers_sorted['txn_per_panelist_yoy'] = reported_numbers_sorted.groupby('entity')['txn_per_panelist'].pct_change(periods=4)*100
reported_numbers_sorted['spend_per_txn_yoy'] = reported_numbers_sorted.groupby('entity')['spend_per_txn'].pct_change(periods=4)*100
reported_numbers_sorted['total_num_panelists_yoy'] = reported_numbers_sorted.groupby('entity')['total_num_panelists'].pct_change(periods=4)*100

cleaned_reported_numbers_sorted = reported_numbers_sorted.dropna(subset=['spend_per_panelist_yoy', 'txn_per_panelist_yoy'])

cleaned_reported_numbers_sorted
```

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error

def fit_regression_models_and_predict(data, entity):
    print(f"--- Regression Models for {entity} ---")
    
    entity_data = data[data['entity'] == entity]
    prediction_data = entity_data[entity_data['reported_sales_yoy'].isnull()]
    training_data = entity_data[entity_data['reported_sales_yoy'].notnull()]
    y_train = training_data['reported_sales_yoy']
    
    # Model 1: spend_per_panelist_yoy, txn_per_panelist_yoy, spend_per_txn_yoy
    X1_train = training_data[['spend_per_panelist_yoy', 'txn_per_panelist_yoy', 'spend_per_txn_yoy']]
    X1_train = sm.add_constant(X1_train)
    model_1 = sm.OLS(y_train, X1_train).fit()
    
    # Model 2: spend_per_panelist_yoy, txn_per_panelist_yoy
    X2_train = training_data[['spend_per_panelist_yoy', 'txn_per_panelist_yoy']]
    X2_train = sm.add_constant(X2_train)
    model_2 = sm.OLS(y_train, X2_train).fit()
    
    # Model 3: spend_per_panelist_yoy
    X3_train = training_data[['spend_per_panelist_yoy']]
    X3_train = sm.add_constant(X3_train)
    model_3 = sm.OLS(y_train, X3_train).fit()
    
    # Model 4: txn_per_panelist_yoy
    X4_train = training_data[['txn_per_panelist_yoy']]
    X4_train = sm.add_constant(X4_train)
    model_4 = sm.OLS(y_train, X4_train).fit()
    
    # Model 5: spend_per_txn_yoy
    X5_train = training_data[['spend_per_txn_yoy']]
    X5_train = sm.add_constant(X5_train)
    model_5 = sm.OLS(y_train, X5_train).fit()
    
    models = [model_1, model_2, model_3, model_4, model_5]
    features = [
        ['spend_per_panelist_yoy', 'txn_per_panelist_yoy', 'spend_per_txn_yoy'],
        ['spend_per_panelist_yoy', 'txn_per_panelist_yoy'],
        ['spend_per_panelist_yoy'],
        ['txn_per_panelist_yoy'],
        ['spend_per_txn_yoy']
    ]
    model_names = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5']
    
    # Predictions for the missing reported_sales_yoy
    predictions = []
    
    for i, model in enumerate(models):
        print(f"\n--- {model_names[i]} ---")
        print(model.summary())  # Display the summary (p-values, R-squared, etc.)
        print(f"R-squared: {model.rsquared}")
        print(f"MSE: {mean_squared_error(y_train, model.fittedvalues)}")
        print("\n")
        X_pred = prediction_data[features[i]]
        X_pred = sm.add_constant(X_pred, has_constant='add') 
        y_pred = model.predict(X_pred)
        predictions.append(y_pred)
    print(f"Predictions for {entity} (Missing reported_sales_yoy):")
    for i, pred in enumerate(predictions):
        print(f"{model_names[i]}: {pred.values}")
    print("\n")
    
for restaurant in cleaned_reported_numbers_sorted['entity'].unique():
    fit_regression_models_and_predict(cleaned_reported_numbers_sorted, restaurant)
```

```python
#plots side by side

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('final yoy.xlsx')

df.loc[(df['entity'] == 'EAT') & df['reported_sales_yoy'].isnull(), 'reported_sales_yoy'] = -0.1384
df.loc[(df['entity'] == 'DRI') & df['reported_sales_yoy'].isnull(), 'reported_sales_yoy'] = -0.0271
df.loc[(df['entity'] == 'TXRH') & df['reported_sales_yoy'].isnull(), 'reported_sales_yoy'] = 0.0763

df['reported_sales_yoy'] = df['reported_sales_yoy'] * 100
df['period_end_dt'] = pd.to_datetime(df['period_end_dt'])

fig, axes = plt.subplots(1, 3, figsize=(24, 6)) 
restaurants = ['TXRH', 'EAT', 'DRI']
axes_titles = {'TXRH': 'Texas Roadhouse Inc.', 'EAT': 'Brinker International, Inc.', 'DRI': 'Darden Restaurants, Inc.'}
colors = {'spend_per_panelist_yoy': 'blue', 'txn_per_panelist_yoy': 'green', 'reported_sales_yoy': 'orange'}

for i, restaurant in enumerate(restaurants):
    restaurant_data = df[df['entity'] == restaurant]
    axes[i].plot(restaurant_data['period_end_dt'], restaurant_data['spend_per_panelist_yoy'], label='Spend Per Panelist YoY', color=colors['spend_per_panelist_yoy'])
    axes[i].plot(restaurant_data['period_end_dt'], restaurant_data['txn_per_panelist_yoy'], label='Transactions Per Panelist YoY', color=colors['txn_per_panelist_yoy'])
    axes[i].plot(restaurant_data['period_end_dt'], restaurant_data['reported_sales_yoy'], label='Reported Sales YoY', color=colors['reported_sales_yoy'])
    axes[i].set_title(f'{axes_titles[restaurant]}: YoY Growth Comparison')
    axes[i].set_xlabel('Period End Date')
    axes[i].set_ylabel('YoY Growth (%)')
    axes[i].legend(loc='best')
    axes[i].grid(True)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('yoy_growth_comparison_restaurants_side_by_side.png', dpi=300)
plt.show()

df.to_excel('updated_final_yoy_data.xlsx', index=False)
```
